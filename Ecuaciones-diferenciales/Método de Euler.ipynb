# Problema 1
Para las ecuaciones que siguen, formar una tabla de los valores de la solución que satisface a la condición inicial dada en el segmento indicado:


1. $$y'=\frac{y}{x}\hspace{1cm}y(1)=1\hspace{1cm}[1,4]\hspace{1cm}(h=0.5)$$

```python
def f(x,y):
    return y/x


h = 0.1
a = 1
b = 4
n = (b-a)/h


x = np.arange(a,b+h, h)
y = np.ones(int(n)+1)

for y_0 in [1, 1.2, 1.4, 1.6, 1.8, 2]:
    y[0] = y_0

    for i in range(0,int(n)):
        y[i+1]= y[i]+f(x[i],y[i])*h
    plt.plot(x,y,label="Aproximación $y_0$"+str(N(y_0, digits=2)))
    plt.legend()
```

2. $$y'=\frac{1}{2}xy\hspace{1cm}y(0)=1\hspace{1cm}[0,1]\hspace{1cm}(h=0.1)$$

```python
def f(x,y):
    return 1/2*x*y

h = 0.1
a = 0
b = 1
n = (b-a)/h


x = np.arange(a,b+h, h)
y = np.ones(int(n)+1)

for y_0 in [1, 1.2, 1.4, 1.6, 1.8, 2]:
    y[0] = y_0

    for i in range(0,int(n)):
        y[i+1]= y[i]+f(x[i],y[i])*h
    plt.plot(x,y,label="Aproximación $y_0$"+str(N(y_0, digits=2)))
    plt.legend()
```
